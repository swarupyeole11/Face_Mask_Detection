No, the shapes (2215,) and (2215,1) are not the same in numpy array.

The shape (2215,) indicates a 1-dimensional numpy array with 2215 elements. This is sometimes called a "flat" array or a "vector". Each element in the array can be accessed with a single index.

On the other hand, the shape (2215,1) indicates a 2-dimensional numpy array with 2215 rows and 1 column. This is sometimes called a "column vector". Each element in the array requires two indices to access, one for the row and one for the column.

Even though both shapes have the same number of elements, they represent different types of data structures and may require different approaches for certain operations in numpy.

```python
#converting 1 D numpy array into 2D numpy array
x_new = x.reshape(2215,1)
print(x_new.shape)

#the same can be achieved using this code
x_new_alpha = x.reshape(-1,1)
# it tells the python to make sure that the data had 1 column and adjust he number of rows for the data accordingly
```

```python
# converting a 2d array in to 3d array
x_3d = x.reshape(1,5,-1)  #make sure when you reshape an array the new dimesions when multpied are equal to the old ones when multpled for e.g     1*2215 = 1*5*443
print(x_3d.shape)
print(x_3d)
```
